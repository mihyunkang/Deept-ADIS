```python
import cv2
import numpy as np
import face_recognition
import re
import os
from PIL import Image

# 한 디렉토리 안의 모든 비디오 파일들에 대해.
def directory_video(directory):
    #frame_save = videoname+'_frame'
    #print(frame_save)
    
    video_files = os.listdir(directory)
    for file in video_files:
        extract_frame(directory+'\\'+file)
```

```python

# 비디오에 대해 프레임 추출
#frame_save_path = videoname+'_frame'

#total = 0

def extract_frame(filename):
    #path 
    print(filename)
    #print("number of video :{}".format(total))
    folder_path = 'D:\\dataset\\real'
    file_name = filename.split('\\')[2]
    video = cv2.VideoCapture(filename)
    count = 0
    num=-1

    #create save folders
    #createfolder(folder_path)
    
    videoframe = video.get(cv2.CAP_PROP_POS_FRAMES)
    videocount = video.get(cv2.CAP_PROP_FRAME_COUNT)

    while video.isOpened():
        #ret 은 프레임이 존재하지 않을때 T/F 반환
        #frame 은 프레임 반환
        num+=1
        '''
        if(videoframe == videocount):
            video.open(filename)
        '''
        ret, frame = video.read()
        frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)    

        if ret == False:
            print("Frame doesn't Exist")
            break
        if num % 100 == 0:
            #total += 1
            if(videoframe == videocount):
                video.open(filename)

            try:

                #face_location 은 얼굴 위치(행렬에서)
                face_location = face_recognition.face_locations(frame) #cnn 모델 사용 X. 원할시, model='cnn' 입력(GPU 없는 기기에서는 실행되지 않음)
                top, right, bottom, left = face_location[0]
                print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

                #face_location 찾아서 파일로 저장
                face_image = Image.fromarray(frame[top-10:bottom+10, left-10:right+10])
                face_image.save(folder_path+'\\'+file_name[:-4]+' %d.jpg'%count)
                print('Saved frame%d.jpg' %count)
                count += 1

            except Exception as e:
                print("Exception Occurs.", e)
                pass

    video.release()
    cv2.destroyAllWindows()
```

```python
#얼굴 탐지
def faceDetect(path, videoname, idnum):
    eye_detect = False
    face_cascade = cv2.CascadeClassifier('haarcascade_frontface.xml')
    eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
    info = ''
    count = 0
    
    try:
        video = cv2.VideoCapture(path)
    except:
        print('video loading failed.')
        return
    
    while True:
        ret, frame = video.read()
        if not ret:
            break
        if eye_detect:
            info = "Eye detection on"
        else:
            info = "Eye detection off"
        
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = face_cascade.detectMultiScale(gray, 1.3,5)
        #top, right, bottom, left = faces
        print(faces)
        #for (x,y,w,h) in faces:
        print("A face is located at pixel location Top: {}, Left: {}, Bottom: {}, Right: {}".format(top, left, bottom, right))

        #face_location 찾아서 파일로 저장
        face_image = Image.fromarray(frame[top:bottom, left:right])
        #Image.fromarray(face_image).show()
        face_image.save(videoname+'/'+idnum+'frame %d.jpg' %count)
        print('Saved frame%d.jpg' %count)
        count += 1
        
        
```

```python
#cascade 분류기를 사용한 얼굴인식

import cv2
import numpy as np
import face_recognition
import re
from matplotlib import pyplot as plt

dirname = 'deepfake-data_subset1/original_videos/'
videoname = '643049'
classnum = '_D_001'
video_path = dirname + videoname+'/'
frame_save_path = videoname+'/'+classnum+'_frame'
path = video_path+videoname+classnum+'.mp4'
#'C:\Users\user\Documents\cyber17\Face_Detection\'
video = cv2.VideoCapture(path)

videoframe = video.get(cv2.CAP_PROP_POS_FRAMES)
videocount = video.get(cv2.CAP_PROP_FRAME_COUNT)

createfolder(frame_save_path)

while video.isOpened():
    #ret 은 프레임이 존재하지 않을때 T/F 반환
    #frame 은 프레임 반환
    if(videoframe == videocount):
        video.open(path)
    
    faceDetect(path, videoname,classnum)


video.release()
cv2.destroyAllWindows()
```
